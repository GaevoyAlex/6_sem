```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset_url = "D:/test_rogects/IAD\Lab3/CAR DETAILS FROM CAR DEKHO.csv"
dataset = pd.read_csv(dataset_url)

dataset[:5]
```

Столбчатая диаграмма

```python
plt.figure(figsize=(20, 12))
sns.barplot(x="name", y="selling_price", data=dataset)
plt.xticks(rotation=90)
plt.xlabel("Марка автомобиля")
plt.ylabel("Средняя цена продажи")
plt.title("Средняя цена продажи по марке автомобиля")
plt.show()
```

Гистограммa

```python
plt.figure(figsize=(8, 6))
sns.histplot(dataset["selling_price"], bins=20, kde=True)
plt.xlabel("Цена продажи")
plt.ylabel("Частота")
plt.title("Распределение цен продажи")
plt.show()
```

Линейный график

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x="year", y="selling_price", data=dataset)
plt.xlabel("Год")
plt.ylabel("Цена продажи")
plt.title("Изменение цены продажи с течением времени")
plt.show()
```

Диаграмма размаха

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="fuel", y="selling_price", data=dataset)
plt.xlabel("Тип топлива")
plt.ylabel("Цена продажи")
plt.title("Распределение цен продажи по типу топлива")
plt.show()
```

Ящик с усами

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x="transmission", y="selling_price", data=dataset)
plt.xlabel("Трансмиссия")
plt.ylabel("Цена продажи")
plt.title("Распределение цен продажи по типу трансмиссии")
plt.show()
```

Радиальная диаграмма

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x="year", y="selling_price", data=dataset, ci=None)
plt.xlabel("Год")
plt.ylabel("Цена продажи")
plt.title("Распределение цен продажи по годам")
plt.show()
```

```python
weather_data_url = "D:/test_rogects/IAD\Lab3/temperature.csv"
weather_df = pd.read_csv(weather_data_url)

weather_df.head()
```

```python
selected_columns = ["Vancouver", "Portland", "San Francisco","Seattle","Los Angeles","San Diego","Las Vegas","Phoenix","Albuquerque","Philadelphia","New York","Montreal", "Boston","Beersheba","Tel Aviv District"	,"Eilat","Haifa","Nahariyya","Jerusalem"]
subset_data = weather_df[selected_columns]

correlation_matrix = subset_data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Корреляция между параметрами погоды")
plt.show()
```

```python
import pandas as pd
import folium

data = pd.read_csv("D:/test_rogects/IAD\Lab3/globalterrorismdb_0718dist.csv", encoding='ISO-8859-1')

data_cleaned = data.dropna(subset=["latitude", "longitude"])

world_map = folium.Map(location=[0, 0], zoom_start=2)

for index, row in data_cleaned.iterrows():
    folium.Marker([row['latitude'], row['longitude']], popup=row['attacktype1_txt']).add_to(world_map)

world_map.save("world_terrorist_attacks_map.html")

```
