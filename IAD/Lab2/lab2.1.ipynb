```python
def determine_measurement_scale(data_list):
    for item in data_list:
        if isinstance(item, str):
            print(f"{item} - номинативная шкала (категориальный признак)")
        elif isinstance(item, int):
            print(f"{item} - порядковая или интервальная шкала")
        elif isinstance(item, float):
            print(f"{item} - интервальная или шкала отношений")
        else:
            print(f"{item} - неизвестный тип данных")

# Пример использования функции
my_data_list = [25, 'мужской', 'высшее образование', 50000.0,25, 'женский', 'среднее образование', 30000.0,30, 'женский', 'среднее образование', 90000.0,25, 'мужской', 'среднее образование', 10000.0]
determine_measurement_scale(my_data_list)

```

```python
import pandas as pd

def determine_measurement_scale(dataset_path):
    
    df = pd.read_csv(dataset_path)

    # Проходим по каждому столбцу
    for column in df.columns:
        unique_values = df[column].nunique()

        # Определяем тип шкалы измерения
        if unique_values == 2:
            print(f"Столбец '{column}' - номинативная шкала (бинарный признак)")
        elif unique_values > 2 and df[column].dtype == 'object':
            print(f"Столбец '{column}' - номинативная шкала (категориальный признак)")
        elif unique_values > 2 and df[column].dtype != 'object':
            print(f"Столбец '{column}' - порядковая или интервальная шкала")
        else:
            print(f"Столбец '{column}' - отношения (непрерывный признак)")



# Пример использования функции
titanic_dataset_path = "D:/test_rogects/IAD\Lab2/titanic.csv"
determine_measurement_scale(titanic_dataset_path)
```

```python
import pandas as pd

titanic_dataset_path = "D:/test_rogects/IAD\Lab2/titanic.csv"
df = pd.read_csv(titanic_dataset_path)

# Мода для пола
gender_mode = df['Sex'].mode().values[0]

# Медиана для возраста
age_median = df['Age'].median()

# Порядок класса билета
pclass_order = df['Pclass'].unique()

# Среднее значение цены билета
fare_mean = df['Fare'].mean()

print(f"Мода для пола: {gender_mode}")
print(f"Медиана для возраста: {age_median}")
print(f"Порядок класса билета: {pclass_order}")
print(f"Среднее значение цены билета: {fare_mean}")


```

```python
import matplotlib.pyplot as plt

age_data = df['Age'].dropna()
gender_data = df['Sex'].value_counts()
fare_data = df['Fare'].dropna()

# Создаем столбчатую диаграмму
plt.bar(gender_data.index, gender_data.values)
plt.xlabel('Пол')
plt.ylabel('Количество пассажиров')
plt.title('Распределение пола пассажиров на "Титанике"')
plt.show()

plt.hist(age_data, bins=20, edgecolor='black')
plt.xlabel('Возраст')
plt.ylabel('Количество пассажиров')
plt.title('Распределение возраста пассажиров на "Титанике"')
plt.show()

plt.hist(fare_data, bins=20, edgecolor='black')
plt.xlabel('Цена билета')
plt.ylabel('Количество пассажиров')
plt.title('Распределение цены билетов на "Титанике"')
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Создание случайных данных (замените их на свои реальные данные)
np.random.seed(42)
data1 = np.random.normal(loc=0, scale=1, size=100)
data2 = np.random.normal(loc=0.5, scale=1, size=100)

# Проведение t-теста
t_statistic, p_value = stats.ttest_ind(data1, data2)

# Визуализация данных
plt.figure(figsize=(8, 6))
plt.hist(data1, bins=20, alpha=0.5, label='Ряд данных 1')
plt.hist(data2, bins=20, alpha=0.5, label='Ряд данных 2')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.title('Распределение двух рядов данных')
plt.legend()
plt.show()

# Вывод результатов t-теста
if p_value < 0.05:
    print(f"p-значение: {p_value:.4f} (статистически значимые различия)")
else:
    print(f"p-значение: {p_value:.4f} (нет статистически значимых различий)")
```
